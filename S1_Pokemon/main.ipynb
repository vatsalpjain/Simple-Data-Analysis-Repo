```python
special = df.groupby(['Type 1','Generation'])
summary_df = special.agg(
    Pokemon_Count=pd.NamedAgg(column='Name', aggfunc='count'),
    Average_Total=pd.NamedAgg(column='Total', aggfunc='mean')
).reset_index()
print(summary_df)
```

```python
import matplotlib.pyplot as plt
generation_counts = df['Type 1'].value_counts().sort_index()
generation_counts
# Plot a bar chart
generation_counts.plot(kind='bar', color='skyblue')

plt.title('Number of Pokmon per Type 1')
plt.xlabel('Type 1')
plt.ylabel('Count of Pokmon')
plt.show()
```
